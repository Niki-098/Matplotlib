```python
from sklearn import datasets
import matplotlib.pyplot as plt
iris = datasets.load_iris()      #Loading the iris dataset
X_iris = iris.data               #Splitting the data into dependent and independent variables
Y_iris = iris.target
average = X_iris[Y_iris == 0].mean(axis=0)


plt.bar(iris.feature_names, average)  #Plotting a bar chart
plt.title("Bar Chart Setosa Averages")
plt.ylabel("Average (in cm)")
plt.show()



```

```python
#Publication quality for clear image after zooming in

import numpy as np
import matplotlib.pyplot as plt


# Create a figure of size 15x12 inches, 1500 dots per inch
plt.figure(figsize=(6, 6), dpi=1500)


var_on_x = np.linspace(-np.pi, np.pi, 256, endpoint=True)


my_cos = np.cos(var_on_x)
var_on_y_sin= np.sin(var_on_x)


plt.plot(var_on_x, my_cos)
plt.plot(var_on_x, var_on_y_sin)


plt.show()



```
