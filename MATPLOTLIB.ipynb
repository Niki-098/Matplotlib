```python
import matplotlib.pyplot as plt
import numpy as np


# Generate some random data
#np.random.seed(0)
x = np.random.randn(100)  # Generate 100 random samples for the x-axis
y = np.random.randn(100)  # Generate 100 random samples for the y-axis
colors = np.random.rand(100)  # Generate random colors for each data point


# Create a scatter plot
plt.scatter(x, y, c=colors, cmap='viridis')  # 'c' specifies the color of each data point
plt.colorbar()  # Add a colorbar for reference


# Set labels and title
plt.xlabel('X-axis')  # Set the x-axis label
plt.ylabel('Y-axis')  # Set the y-axis label
plt.title('Scatter Plot of Random Data')  # Set the title of the scatter plot


# Show the scatter plot
plt.show()





```

```python
import matplotlib.pyplot as plt


# Define the data
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']  # Categories/labels for each slice of the pie chart
sizes = [99, 30, 20, 15]  # Sizes/values for each slice of the pie chart
colors = ['red', 'yellow', 'orange', 'green']  # Colors for each slice of the pie chart


# Create a pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)


# Set title
plt.title('Pie Chart of Fruit Distribution')


# Show the pie chart
plt.show()





```

```python
import matplotlib.pyplot as plt
import numpy as np


# Generate some random data
np.random.seed(0)
n_points = 100
x = np.linspace(-5, 5, n_points)
y = np.linspace(-5, 5, n_points)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**10 + Y**10))


# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')


# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Plot of Function')


# Show the 3D plot
plt.show()





```

```python
import matplotlib.pyplot as plt
import numpy as np


# Generate some random data
np.random.seed(0)
n_points = 100
x = np.linspace(-5, 5, n_points)
y = np.linspace(-5, 5, n_points)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 - Y**2))


# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')


# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Plot of Function')


# Show the 3D plot
plt.show()





```
